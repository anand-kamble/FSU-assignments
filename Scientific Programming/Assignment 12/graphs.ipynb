```python
import matplotlib.pyplot as plt

# Data
threads = [1, 2, 4, 8]
time_taken = [0.265110, 0.146537, 0.086372, 0.055592]
speedup = [1.000000, 1.809186, 3.069453, 4.769009]

# Plotting Time taken
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(threads, time_taken, marker='o', linestyle='-', color='b')
plt.title('Time taken vs No. of threads')
plt.xlabel('No. of threads')
plt.ylabel('Time taken')
plt.grid(True)

# Plotting Speedup
plt.subplot(1, 2, 2)
plt.plot(threads, speedup, marker='o', linestyle='-', color='r')
plt.title('Speedup vs No. of threads')
plt.xlabel('No. of threads')
plt.ylabel('Speedup')
plt.grid(True)

# Adjust layout for better appearance
plt.tight_layout()

# Show the plot
plt.show()
```
