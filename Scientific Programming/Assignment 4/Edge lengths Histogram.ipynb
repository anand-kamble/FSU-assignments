```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Import the file
file = open("edge_lengths.txt", "r")

# Read the contents of the file into a string
string = file.read()

# Close the file
file.close()
```

```python
string = string.split("\n")
```

```python
for x in string:
    try:
        float(x) 
    except ValueError:
        print("x : ",x)
```

```python
lengths = np.array([float(x) for x in string if x != ""])
```

```python
plt.hist(lengths,bins=20)
plt.xlabel("Length")
plt.ylabel("Number of Edges")
plt.savefig('Edge Lengths.png', dpi=300)
```
