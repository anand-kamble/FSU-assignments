```python
import matplotlib.pyplot as plt
import numpy as np
# Sample data (replace this with your actual data)
N_values = np.array([32, 32, 64, 64, 128, 128, 256, 256, 512, 512, 1000, 1000])
quad_tree_times = [61, 65, 187, 246, 724, 990, 2870, 3838, 14225, 15258, 44547, 59977]
brute_force_times= [59, 73, 209, 268, 807, 1047, 3204, 4707, 13105, 17040, 53261, 68698]

# Plotting
plt.plot(N_values[1::2], brute_force_times[1::2], label='Brute Force')
plt.plot(N_values[1::2], quad_tree_times[1::2], label='Quad Tree')

# Add labels and title
plt.xlabel('N')
plt.ylabel('Time ($\mu$s)')
plt.grid()
plt.title('Time vs. N for Brute Force and Quad Tree')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
