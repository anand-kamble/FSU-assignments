```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = [1,2,4,8,16]
y = [4.0, 7.6, 14.9, 23.1, 35.6]
```

```python
# Parallel Efficiency
efficiency = []
for i in range(1,len(x)+1):
    efficiency.append(y[i-1]/(x[i-1]*y[0]))
```

```python
speedup = []
for i in range(1,len(y)+1):
    speedup.append(y[i-1]/y[0])
```

```python
# plt.xscale('log',base=2)
# plt.yscale('log',base=2)
plt.plot(x,x,'r-',linewidth=1)
plt.plot(x,speedup,'o-',linewidth=1)

for i in range(len(x)):
    plt.plot([x[i],x[i]], [0,x[i]], 'k--',alpha=0.3, lw=1)  # Vertical dotted line
    plt.plot([x[i],x[i]], [16,x[i]], 'k--',alpha=0.3, lw=1)  # Vertical dotted line
    plt.plot([x[i], 0], [x[i], x[i]], 'k--',alpha=0.3, lw=1)  # Horizontal dotted line
    plt.plot([x[i], 16], [x[i], x[i]], 'k--',alpha=0.3, lw=1)  # Horizontal dotted line

# plt.yticks(x, x)
# plt.xticks(x, x)
plt.xlabel("No. of Processors")
plt.ylabel("Speedup")
plt.title("Scalability")
plt.show()
```
