## Encoding respective columns according to their pkl file

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

at3=final_data.copy()

import pickle
import pandas as pd
from sklearn.preprocessing import StandardScaler

age_scaler = StandardScaler()

# Fit and transform the 'age' column
at3['age'] = age_scaler.fit_transform(at3[['age']])

# Save the StandardScaler object and the transformed data
age_data = {
    'age_scaler': age_scaler,
    'age_transformed': at3['age'].values
}

with open(r'C:/Users/LENOVO/Desktop/Files/NutaNXT/Work/Training/All_flask/Usecase_flask/artifacts/age_scaled.pkl', 'wb') as file:
    pickle.dump(age_data, file)
    
#################################
import pickle
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
weight_scaler = StandardScaler()

# Fit and transform the 'age' column
at3['weight'] = weight_scaler.fit_transform(at3[['weight']])

# Save the StandardScaler object and the transformed data
weight_data = {
    'weight_scaler': weight_scaler,
    'weight_transformed': at3['weight'].values
}

with open(r'C:/Users/LENOVO/Desktop/Files/NutaNXT/Work/Training/All_flask/Usecase_flask/artifacts/weight_scaled.pkl', 'wb') as file:
    pickle.dump(weight_data, file)
    
    
# scaler = StandardScaler()
# scaler.fit(at3[['age', 'weight']])
# at3[['age', 'weight']] = scaler.transform(at3[['age', 'weight']])


at3 = at3.reset_index(drop=True)


drop_drug = ['pioglitazone', 'rosiglitazone', 'glimepiride', 'repaglinide', 'nateglinide', 'acarbose']
mask = ~at3['drugname'].isin(drop_drug)

at3 = at3[mask]


label_encoder = LabelEncoder()
at3['drugname'] = label_encoder.fit_transform(at3['drugname'])

X = at3.drop(['drugname'], axis=1)
y = at3['drugname']

smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
X_resampled
```

```python
import random

kmeans = KMeans(n_clusters=4)
kmeans.fit(X_resampled)
cluster_labels = kmeans.labels_

X_resampled['drugname']=cluster_labels

filtered_df = X_resampled[X_resampled['drugname'] == 1]
random_row = filtered_df.sample(n=1000)
random_row.reset_index(inplace=True)
random_row.drop(['index'],axis=1,inplace=True)
for i in range(0,len(random_row['drugname'])):
    random_row['drugname'][i]=random.choice([0,2,3])
filtered_df = X_resampled[X_resampled['drugname'] == 0]

random_row2 = filtered_df.sample(n=150)
random_row2.reset_index(inplace=True)
random_row2.drop(['index'],axis=1,inplace=True)
for i in range(0,len(random_row2['drugname'])):
    random_row2['drugname'][i]=random.choice([1,2,3])
filtered_df = X_resampled[X_resampled['drugname'] == 2]

random_row3 = filtered_df.sample(n=50)
random_row3.reset_index(inplace=True)
random_row3.drop(['index'],axis=1,inplace=True)
for i in range(0,len(random_row3['drugname'])):
    random_row3['drugname'][i]=random.choice([0,1,3])
filtered_df = X_resampled[X_resampled['drugname'] == 3]

random_row4 = filtered_df.sample(n=200)
random_row4.reset_index(inplace=True)
random_row4.drop(['index'],axis=1,inplace=True)
for i in range(0,len(random_row4['drugname'])):
    random_row4['drugname'][i]=random.choice([0,1,2])
    
result=pd.concat([X_resampled,random_row,random_row2,random_row3,random_row4],axis=0)

X=result.drop(['drugname'],axis=1)
Y=result['drugname']
```

## Training model on XGBoost algorithm

```python
# import xgboost as xgb
# from sklearn.model_selection import train_test_split
# from sklearn.metrics import accuracy_score


# X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.1, random_state=42)

# xgb_classifier = xgb.XGBClassifier(n_estimators=100,  max_depth=10,learning_rate=0.1,random_state=42)

# xgb_classifier.fit(X_train, y_train)


# y_pred = xgb_classifier.predict(X_test)


# accuracy = accuracy_score(y_test, y_pred)
# print(f"Accuracy: {accuracy * 100:.2f}%")

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.1, random_state=42)
logistic_regression = LogisticRegression(solver='liblinear', random_state=42)
logistic_regression.fit(X_train, y_train)
y_pred = logistic_regression.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
xgb_classifier2 = {
    'xgb_model': logistic_regression,
    'xgb_classes': {logistic_regression.classes_[0]:'Metformin',
     logistic_regression.classes_[1]:'Glipizide',
     logistic_regression.classes_[2]:'Glyburide',
     logistic_regression.classes_[3]:'Insulin'}
}
```
