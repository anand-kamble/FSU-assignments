```python
# Given data
y = np.array([62, 66, 72, 78, 85, 89, 91, 90, 87, 80, 72, 65])
x = np.arange(12) / 12.0  # x = [0/12, 1/12, 2/12, ..., 11/12]

m = 4

A = np.column_stack([np.cos(2 * np.pi * i * x) for i in range(m)])

a, _, _, _ = np.linalg.lstsq(A, y, rcond=None)

display(Latex(f"Coefficients $a_i$ : {a}"))
```

```python
import sympy as sp
import numpy as np

# define x, weight, and limits of integration
x, w, a, b = sp.symbols('x, w, a, b')

        
a = -np.inf                            # lower limit
b = np.inf                             # upper limit
w = lambda x: sp.exp(-x**2)   # weight function(x)


# define inner product, with defaults
def inner_product(f, g, w, a, b):
    return sp.integrate(w*g*f, (x, a, b))


def gram_schmidt(basis):
    orth_basis = []

    for q in basis:
        
        print ('working on monobasis: ',q)
        for prev_q in orth_basis:
            q = q - inner_product(prev_q, q, w(x), a, b)*prev_q
        
#         orthonormalization
        q = q / sp.sqrt(inner_product(q, q, w(x), a, b))

        orth_basis.append(q)
        
        print ('result:',q,'\n')
        
    return orth_basis

# monoBasis = [1, x, x**2]
monoBasis = [1,2*x,4*x**2 -2]
# monoBasis = [1, x, x**2-1]
#monoBasis = [1, x, 6-8*x+x**2+x**3, 4*x-2*x**2+x**3, x+x**2+7*x**4]


orth_basis = gram_schmidt(monoBasis)
# print orthogonal function
for i in range(len(monoBasis)):
    print ("order ",i,' ==> ',orth_basis[i])   
```
