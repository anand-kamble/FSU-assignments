```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def p(x,mu,sigma):
    return 1 / np.sqrt(2 * np.pi * sigma**2)*np.exp((-1/2)*((x-mu)**2/sigma**2))
```

```python
def Finverse(u):
    a0 = 2.30753
    a1 = 0.27061
    b1 = 0.99229
    b2 = 0.04481
    if u <= 0.5:
        t = np.sqrt(-2*np.log(u))
        return (a0+(a1*t)/(1+(b1*t)+(b2*(t**2))))-t
    elif u > 0.5:
        t = np.sqrt(-2*np.log(1-u))
        return t-(a0+(a1*t)/(1+(b1*t)+(b2*(t**2))))
```

```python
numberOfPoints = 10**4

u = np.random.uniform(0,1,size=numberOfPoints)
u = [p(x,0,1) for x in u]
samples = [Finverse(ui) for ui in u]
```

```python
plt.hist(samples);
```

```python
def dxfx(x):
    return -x*np.exp(-(x**2/2))
x_bar = [dxfx(x) for x in samples]
```

```python
plt.hist(x_bar);
```
