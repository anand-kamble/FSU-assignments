# Uniform Sampling of Points inside a Circle

\begin{align*}
\pi(x|y) & = U[-\sqrt{1-y^2}, \sqrt{1-y^2}]\\
\pi(y|x) & = U[-\sqrt{1-x^2}, \sqrt{1-x^2}]\\
\end{align*}

Gibbs Algorithm:

1. Set $i = 0$.
2. Pick a initial state $\mathbf{x}^{(0)} = \{x_1^{(0)}, x_2^{(0)}, x_3^{(0)}\}$
3. Generate the next sample $\mathbf{x}^{(i+1)}$ by sampling each $x_j$, ($j = 1, 2, 3, ... n$) from its conditional PDF
    * sample $x_1^{(i+1)} \sim \pi(x_1 | x_2^{(i)}, x_3^{(i)})$
    * sample $x_2^{(i+1)} \sim \pi(x_2 | x_1^{(i+1)}, x_3^{(i)})$
    * sample $x_3^{(i+1)} \sim \pi(x_3 | x_1^{(i+1)}, x_2^{(i+1)})$
4. Set $\mathbf{x}^{(i+1)} = \{x_1^{(i+1)}, x_2^{(i+1)}, x_3^{(i+1)}\}$
5. Set $i = i + 1$.
6. Go to step 3, and repeat, until enough samples have been collected.


```
npts = 1000
x = np.zeros(npts)
y = np.zeros(npts)

for i in range(1,npts):
    x[i] = np.random.uniform(-np.sqrt(1-y[i-1]**2), np.sqrt(1-y[i-1]**2))
    y[i] = np.random.uniform(-np.sqrt(1-x[i]**2), np.sqrt(1-x[i]**2))


plt.plot(x,y,'o', alpha=0.5)
plt.xlim(-1,1); plt.ylim(-1,1)
plt.xlabel('$x$') 
plt.ylabel('$y$') 
plt.gca().set_aspect('equal')
```

```
plt.hist(x, density=True)
plt.hist(y, density=True)
```
